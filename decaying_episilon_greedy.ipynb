# linear decay

```python
import matplotlib.pyplot as plt

class LinearDecay:
    def __init__(self, initial_value, min_value, decay_rate):
        self.value = initial_value
        self.min_value = min_value
        self.decay_rate = decay_rate

    def decay(self):
        if self.value > self.min_value:
            self.value -= self.decay_rate
        return self.value

decay_schedule = LinearDecay(initial_value=1.0, min_value=0.1, decay_rate=0.01)
values = []

for _ in range(200):
    values.append(decay_schedule.decay())

plt.plot(values)
plt.title("Linear Decay")
plt.xlabel("Time Steps")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

# exponential decay

```python
import matplotlib.pyplot as plt

class ExponentialDecay:
    def __init__(self, initial_value, decay_factor):
        self.value = initial_value
        self.decay_factor = decay_factor

    def decay(self):
        self.value *= self.decay_factor
        return self.value


decay_schedule = ExponentialDecay(initial_value=1.0, decay_factor=0.9)
values = []

for _ in range(20):
    values.append(decay_schedule.decay())

plt.plot(values)
plt.title("Exponential Decay")
plt.xlabel("Time Steps")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

# discrete interval decay

```python
import matplotlib.pyplot as plt

class DiscreteIntervalDecay:
    def __init__(self, initial_value, interval, decay_rate):
        self.value = initial_value
        self.interval = interval
        self.decay_rate = decay_rate
        self.steps = 0

    def decay(self):
        self.steps += 1
        if self.steps % self.interval == 0:
            self.value *= self.decay_rate
        return self.value

    
decay_schedule = DiscreteIntervalDecay(initial_value=1.0, interval=3, decay_rate=0.5)
values = []

for _ in range(20):
    values.append(decay_schedule.decay())

plt.plot(values)
plt.title("Discrete Interval Decay")
plt.xlabel("Time Steps")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
